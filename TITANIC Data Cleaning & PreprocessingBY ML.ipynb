```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Drop columns with too many missing values
df.drop(columns=['Cabin'], inplace=True)

# Fill missing Age with median (more robust to outliers than mean)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Embarked with mode (most common port)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Convert categorical features into numerical using encoding
# Convert 'Sex' and 'Embarked' to dummy variables
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

# Drop 'Name' and 'Ticket' for now
df = df.drop(columns=['Name', 'Ticket'])
```

```python
# Normalize/standardize the numerical features
from sklearn.preprocessing import StandardScaler

# Standardize Age and Fare
scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
# Boxplot to see outliers
plt.figure(figsize=(12, 5))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Outlier Visualization")
plt.show()

# Remove outliers in 'Fare' using IQR
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Fare'] < (Q1 - 1.5 * IQR)) | (df['Fare'] > (Q3 + 1.5 * IQR)))]
```
